```python
import matplotlib
matplotlib.rcParams.update({'font.size': 18})

fig, ax = plt.subplots(2,2, figsize=(15,10))
ax = ax.ravel()
df = pd.read_csv('validation_curves.txt')
df.plot('alpha', 'f1_score_0', label='Model 0', ax=ax[0])
df.plot('alpha', 'f1_score_1', ax=ax[0], label='Model 1')
df.plot('alpha', 'f1_score_2', ax=ax[0], label='Model 2')
ax[0].set_ylabel('F1, weighted')
ax[0].set_title('Validation curves')
ax[0].set_xscale('log')

for i, j, a in zip(['learning_curve_0.txt', 'learning_curve_1.txt', 'learning_curve_2.txt'], 
    ['Model 0', 'Model 1', 'Model 2'], ax[1:]):
    df = pd.read_csv(i, delim_whitespace=True)
    df.plot('N', 'train_f1_score', label='Train', ax=a)
    df.plot('N', 'test_f1_score', ax=a, label='Test')
    a.set_title(f'Learning curves ({j})')
    a.set_ylabel('F1, weighted')
    a.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.tight_layout()
plt.savefig('f1_curves.png')
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,10))
ax = ax.ravel()
df = pd.read_csv('validation_curves.txt')
df.plot('alpha', 'accuracy_0', label='Model 0', ax=ax[0])
df.plot('alpha', 'accuracy_1', ax=ax[0], label='Model 1')
df.plot('alpha', 'accuracy_2', ax=ax[0], label='Model 2')
ax[0].set_ylabel('Mean accuracy')
ax[0].set_title('Validation curves')
ax[0].set_xscale('log')

for i, j, a in zip(['learning_curve_0.txt', 'learning_curve_1.txt', 'learning_curve_2.txt'], 
    ['Model 0', 'Model 1', 'Model 2'], ax[1:]):
    df = pd.read_csv(i, delim_whitespace=True)
    df.plot('N', 'train_score', label='Train', ax=a)
    df.plot('N', 'test_score', ax=a, label='Test')
    a.set_title(f'Learning curves ({j})')
    a.set_ylabel('Mean accuracy')
    a.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.tight_layout()
plt.savefig('accuracy.png')
```
