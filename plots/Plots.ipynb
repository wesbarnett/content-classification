```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('validation_curves.txt')
```

```python
df.head()
```

```python
ax = df.plot('alpha', 'f1_score_0', label='Popular')
ax = df.plot('alpha', 'f1_score_1', ax=ax, label='Kinda popular')
df.plot('alpha', 'f1_score_2', ax=ax, label='Not-as-popular')
plt.ylabel('F1, weighted')
plt.title('Validation curves')
plt.xscale('log')
plt.ylim([0.4,0.7])
plt.savefig('validation_curves.png')
```

```python
ax = df.plot('alpha', 'accuracy_0')
ax = df.plot('alpha', 'accuracy_1', ax=ax)
df.plot('alpha', 'accuracy_2', ax=ax)
plt.ylabel('Validation score')
plt.title('Validation curves')
plt.xscale('log')
```

```python
for i, j in zip(['learning_curve_2.txt', 'learning_curve_1.txt', 'learning_curve_0.txt'], 
    ['Not-as-popular', 'Kinda popular', 'Popular']):
    df = pd.read_csv(i, delim_whitespace=True)
    ax = df.plot('N', 'train_f1_score', label='Train, F1 (weighted)')
    df.plot('N', 'test_f1_score', ax=ax, label='Test, F1 (weighted)')
    plt.title(f'Learning curves ({j})')
```

```python
for i, j in zip(['learning_curve_2.txt', 'learning_curve_1.txt', 'learning_curve_0.txt'], 
    ['Small', 'Medium', 'Large']):
    df = pd.read_csv(i, delim_whitespace=True)
    ax = df.plot('N', 'train_score')
    df.plot('N', 'test_score', ax=ax)
    plt.title(f'Learning curves ({j})')
```
