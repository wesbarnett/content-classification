```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('validation_curves.txt', delim_whitespace=True)
ax = df.plot('alpha', 'f1_score_0')
ax = df.plot('alpha', 'f1_score_1', ax=ax)
df.plot('alpha', 'f1_score_2', ax=ax)
plt.ylabel('Validation score')
plt.title('Validation curves')
plt.xscale('log')
```

```python
df = pd.read_csv('validation_curves.txt', delim_whitespace=True)
ax = df.plot('alpha', 'accuracy_0')
ax = df.plot('alpha', 'accuracy_1', ax=ax)
df.plot('alpha', 'accuracy_2', ax=ax)
plt.ylabel('Validation score')
plt.title('Validation curves')
plt.xscale('log')
```

```python
for i, j in zip(['learning_curve_2.txt', 'learning_curve_1.txt', 'learning_curve_0.txt'], 
    ['Small', 'Medium', 'Large']):
    df = pd.read_csv(i, delim_whitespace=True)
    ax = df.plot('N', 'train_f1_score')
    df.plot('N', 'test_f1_score', ax=ax)
    plt.title(f'Learning curves ({j})')
```
